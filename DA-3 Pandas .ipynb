# Zip function data frame


```python
list_a = ['a','b','c','d','e']
list_b = [1,2,3,4,5]
df = pd.DataFrame(zip(list_a,list_b),index = [1,2,3,4,5],columns = ['Alphabets','Position'])
display(df)
```

**Let's undertand how it works**
so zip function basically commbines elements form list_a and list_b together in tuple it does this element by element i.e (a,1),(b,2)...... and return this new pair of elements in list ir whatever data type user defines 

```python
zipped = list(zip(list_a,list_b))
zipped
```

**The output [('a', 1), ('b', 2), ('c', 3), ('d', 4), ('e', 5)] kind of looks similar to 2-d array and pandas treats it likewise.**
