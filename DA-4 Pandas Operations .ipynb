Lets talk about lambda functions. So before doing that lets revisit the function/method part, We can consider loosely that the function and methods as to be same. So in order to define the functions we wound need a name for the function, parameter, and block of code to perform some action and return the desired output 

**Lambda functions on the other hand are anonymous functions which means that you don't have to name the functions so there goes the naming part you are now left with paratmeters so in lambda funciotns it can have n numbers of arguments and coming to the code block, there is one issue here in lambda functions you are allowed to write only one expression (the line of code which is evaluated , where as statement is just stating), and it returns only one output**

**A more formal defenition of a lambda function would be it is an anonymous function which can take any number of arguments but can have 
one expression, arguments is parameter here and expression is the line of code so it can take n parmeteres but can have one line of code only** 

```python
y = lambda a,b,c,d,e: a+b+c+d+e
y(1,2,3,4,5)
```

```python
def profit_percentage(column):
    
    try:
        if column[2]<column[22]:
            pp = ((column[22] -column[2])/column[2])*100
            roundno = round(pp,2)
            return roundno
        
        else:
            pp = ((column[2] -column[22])/column[2])*100
            roundno = round(pp,2)
            return roundno
    
    except:
        nan = np.nan
        return nan
    
    else:
        profit_per = round(pp,2)
        return profit_per
        
# this is not the way we write function generally this is some what customed for pandas
```

## Treating null values in dataframe


```python
# We will come across null value in the data frame hence it becomes very important to know how to deal with such problems 
# first this is to look for null values 
df['homepage'].isnull()
# This is basically a boolean operations it return true or false if the there is or not a null value
```

lets understand the output True means yes the value at that position is null i.e zero 

Fale means no the value at that postion is not zero 

```python
# Let's count the sum of null values in the homepage column
df['homepage'].isnull().sum()
# The output tells us that there are 2054 values in the column that are null values 
```

```python
print((~df['homepage'].isnull).sum()) # give the sum of complement of the null values i.e is notnull values
# can also be print as
print(df.homepage.notnull().sum())
```

## what if we want all the non null values than we will have to do the negative of the above line of code for this purpose tilda (~) is used 


```python
df[~df['homepage'].isnull()]
#this gives all non null values 
```

```python
#lets find the null values for this as well 
df[~df['homepage'].isnull()].sum() # can not do it for this line of code 
```
