## Inner join

```python
inner_join=pd.merge(Table_A,Table_B,how = 'inner',left_on ='C',right_on = 'C')
display(inner_join)
display(Table_A)
display(Table_B)
```

So in Table_A and Table_B, column "C" is common and we are doing inner join on column "C", How excatly will the inner join come about. It will take all the common elements of column C from Table_A and Table_B and create a new table. The column C of new table will have only the common elements which are present in Table_A and Table_B , and the rest of the column i.e A, B, D, E in the new table will have only those values which correspond to the common elements in C.

Note: if the common elements in the C are not unique that is we have two 7,s in column C in table A and two 7's in column C in table B then there comes the mapping concept i.e one 7 in Table_a will be mapped to the two 7's and in Table_B, the other 7 in Table_A will be map with the 2 7's in Table_B, it is just like we did in maths function and relation class

```python
# lets do the inner join for the other table and see what happens
inner_join_2 = pd.merge(Table_X,Table_Y,how = 'inner', left_on = 'Gamma', right_on = 'Gamma')
display(inner_join_2)
display(Table_X)
display(Table_Y)
```

What excatly happends in outer join is the that it takes all the rows from the right table and all the rows from the left table and it clubs the table on the column columns.
