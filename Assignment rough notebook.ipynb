```python
class cylinder:
    def __init__(self,radius,height): 
        self.radius = radius #  the self is used to put the attributes in the memory of class
        self.height = height # we are creating a vraible and telling it to remember 
        
    def volume(self):
        vol = 3.14*(self.radius)**2*self.height
        return vol
        
    def surface_area(self):
        sa = 2*3.14*self.radius*self.height
        return sa
            
```

```python
class BankAccount:
    def __init__(self,balance = 0):
        self.balance = balance
        
        
    def withdraw(self,amount): # apart form the already parameter you can take aditional parameters as well
        if amount <= self.balance:
            self.balance = self.balance - amount
            #print(f'Amount {amount} is deducted from your account\nYou have {self.balance} left in your account' )
            return self.balance
        else:
            y = 'You don\'t have enough balance to do this transaction'
            return y
    
    def deposit(self,amount):
        self.balance = self.balance + amount
        #print(f'Amount {amount} has been credited in your account\nYou have {self.amount} in your account')
        return self.balance
    
# the use of self is very important here because every time a change in balance is there then it should take the new change
# this is kind of reasigning hence we are using reasigning here
```

```python
class Time:
    def __init__(self,hours,minutes):
        self.hours = hours
        self.minutes = minutes
        
    def displaytime(self):
        if self.hours <= 11 and self.minutes<= 59:
            y='AM'
            formatt =(f'{self.hours}:{self.minutes} AM')
            return formatt
        elif self.hours == 12 and self.minutes <= 59:
            y = 'PM'
            formatt = (f'{self.hours}:{self.minutes} PM')
            return formatt
        elif self.hours <= 23 and self.minutes <= 59:
             y = 'PM'
             formatt = (f'{self.hours}:{self.minutes} PM')
             return formatt
        elif self.hours == 24 and self.minutes <= 59:
             formatt = (f'{self.hours}:{self.minutes} PM')
             return formatt
        elif self.hours > 24 or self.minutes > 59: # very good use of or which covers both the case
             y = 'The hours should be less then 24 and the minutes should be less the 60'
             return y
            
    def displayratio(self):
        try:
            ratio = self.minutes/self.hours
        except:
            ratio = 'Cannot divide by zero'
        return ratio
            
            
```
